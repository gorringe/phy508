```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = 40          # grid size MxM
V = 1.0         # top wall Vo
target = 1e-4   # accuracy

phi = np.zeros([M+1,M+1],float)      # previous iteration
phiprime = np.empty([M+1,M+1],float) # current interation
```

```python
delta = 1.0 # iteration dV

while delta > target:
    
    for i in range(M):
        for j in range(M):
            
            if i == 10 and j > 10 and j < 30: # boundary condition left plate
                phiprime[i,j] = +V
            elif i == 30 and j > 10 and j < 30: # boundary condition right plate
                 phiprime[i,j] = -V
            else:
                phiprime[i,j] = (phi[i+1,j] + phi[i-1,j] + phi[i,j+1] + phi[i,j-1])/4

    delta = np.max(abs(phi-phiprime))       # iteration max difference
    phi,phiprime = phiprime,phi
    
```
