```python
import matplotlib.pyplot as plt
import numpy as np

Npi = 100              # pion number
Nmu = 0                # muon number
tau = 26.0             # lifetime of charged pion (ns)

dt = 1.0               # size of time interval (ns)
tmax = 100             # time range (ns)

prob = dt/tau          # decay probability in interval dt

```

```python
time = np.arange(0.0,tmax,dt)
Nplist = []
Nmlist = []

for it in time: # loop over time steps
    Nplist.append(Npi)
    Nmlist.append(Nmu)
    Ndecay = 0
    for ip in range(Npi): # loop over surviving pions
        if np.random.ranf() < prob:
            Ndecay += 1 # count decays in interval dt
    Npi -= Ndecay
    Nmu += Ndecay
```

```python
plt.plot(time,Nplist,"m.")
plt.plot(time,Nmlist,"g.")
plt.xlabel("time [ns]")
plt.ylabel("Npi, Nmu")
```
