```python
import numpy as np
```

```python
# taylor series estimate for exp(x), 1+x+x^2/2!+...

x = 1.0
expseries = 0.0
N = 6

for i in range(N):
    expseries += x**i/np.math.factorial(i)

print("x = ", x,", e^x = ", np.exp(x), ", 1 + x + x^2/2! + ... = ", expseries, ", approx error = ,", (np.exp(x)-expseries)/np.exp(x))
```

```python
# taylor series estimate for cosx, 1-x^2/!+x^4/4!...

x = np.pi/4.
cosseries = 0.0
N = 3

for i in range(N):
  cosseries += ((-1)**i) * (x**(2*i)) / np.math.factorial(2*i)

print("x = ", x, ", cos = ", np.cos(x), ", 1 - x^2/2! + x^4/4! +... = ", cosseries, ", approx error = ,", (np.cos(x)-cosseries)/np.cos(x))
```
