```python
from pylab import np
import matplotlib.pyplot as plt
```

```python
N = 2000 # number of steps
dt = 1.e-2 # step size [s]
tf = dt*float(N) # time interval [s]

vo = 0.0 # initial velocity [m/s]
g = 9.8 # acceleration of gravity [m/s]
coverM = 8.7e-3/10. #C/m  quadratic drag coeff / objects mass [m^-1]
```

```python
def euler(vi, ti, dt, deriv): # Euler method v_i â†’ v_i+1
    viplus1 = vi + deriv(vi,ti) * dt
    return viplus1

def freefall( vi, ti): # calc differential
    dvdt = -g + coverM*vi**2 # gravity down, drag up
    return dvdt
```

```python
t = np.linspace(0, tf, N) # declare time array
v = np.zeros([N]) # declare velocity array

v[0] = vo # initial condition

for j in range(N-1):
    v[j+1] = euler(v[j], t[j], dt, freefall)
```

```python
plt.plot(t,v,"bo",label="numerical")
plt.plot(t,-g*t,"r",label="-gt")
plt.xlabel("time [s]")
plt.ylabel("velocity [m/s]")
plt.legend()
```
