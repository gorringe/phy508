```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = 25                               # grid size MxM
V = 1.0                              # y=0 wall BC, Vo
target = 1e-4                        # tolerance
phi = np.zeros([M+1,M+1],float)      # previous iteration
phiprime = np.empty([M+1,M+1],float) # current interation
phi[:,0] = V                         # boundary condition

xwire,ywire = 12,5                   # wire location
phi[xwire,ywire] = 0                 # grounded wire
```

```python
delta = 1.0 # iteration dV

while delta > target:

    for i in range(M+1):
        for j in range(M+1):

            if i==0 or i==M or j==0 or j==M: # boundary conditions, do not change wall V
                phiprime[i,j] = phi[i,j]
            elif i==xwire and j==ywire:      # boundary conditions, do not change wire V
                phiprime[i,j] = phi[i,j]
            else:
                phiprime[i,j] = (phi[i+1,j] + phi[i-1,j] + phi[i,j+1] + phi[i,j-1])/4

    delta = np.max( abs(phi-phiprime))       # calc, difference array, find max difference
    phi,phiprime = phiprime,phi

plt.imshow(phi, extent=[0,M,0,M], origin='lower') # note it's rotated
```

```python
plt.plot(phi[:, 0],"r-",label="(x, 0)")
plt.plot(phi[:, 1],"g-",label="(x, 1)")
plt.plot(phi[:, 5],"b-",label="(x, 5)")
plt.plot(phi[:,10],"y-",label="(x,10)")
plt.plot(phi[:,15],"m-",label="(x,15)")
plt.plot(phi[:,20],"c-",label="(x,20)")
plt.ylabel("potential at y = 0, 1, 5, 10, 15, 20")
plt.xlabel("position, x")
plt.legend()
```
