```python
from pylab import np
import matplotlib.pyplot as plt
```

```python
N = 400 # number of steps
tau = 2.0 # time interval [s]
dt = tau / float(N-1) # time step [s]
lbda = 1.0 # growth constant [s^-1]
yo = 1.0 # initial population
```

```python
t = np.linspace(0, tau, N) # time array
y = np.zeros([N]) # y0, y1, ... array
```

```python
def euler(yi, ti, dt, deriv): # Euler method y_i+1 â†’ y_i
    ynext = yi + deriv(yi,ti) * dt
    return ynext
```

```python
def growth( yi, ti): # define differential
    dydt = lbda*yi
    return dydt
```

```python
y[0] = yo # initial condition
y[1] = euler(y[0], t[0], dt, growth)
print(y[0],y[1])
```

```python
for j in range(N-1):
    y[j+1] = euler(y[j], t[j], dt, growth)
```

```python
plt.plot(t,y,"bo")
plt.plot(t,np.exp(lbda*t),"r")
```
