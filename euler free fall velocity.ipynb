```python
from pylab import np
import matplotlib.pyplot as plt
```

```python
N = 100 # number of steps
t = 1.0 # free fall time
dt = t / float(N-1) # time step [s]
vo = 0.0 # initial velocity
g = 9.8 # acceleration of gravity
m = 1.0 # mass [kg]
b = 0.5 # drag coefficient
```

```python
t = np.linspace(0, t, N) # time array
v = np.zeros([N]) # v0, v1, ... array
```

```python
def euler(vi, ti, dt, deriv): # Euler method v_i â†’ v_i+1
    vnext = vi + deriv(vi,ti) * dt
    return vnext
```

```python
def freefall( vi, ti): # define differential
    dvdt = -g -b*vi
    return dvdt
```

```python
v[0] = vo # initial condition
v[1] = euler(v[0], t[0], dt, freefall)
```

```python
for j in range(N-1):
    v[j+1] = euler(v[j], t[j], dt, freefall)
```

```python
def freefall( vi, ti): # define differential
    dvdt = - g - b*v
    return dvdt
```

```python
plt.plot(t,v,"bo")
plt.plot(t,-g*t,"r")
```
