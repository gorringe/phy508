```python
from pylab import np
import matplotlib.pyplot as plt
```

```python
N = 2000 # number of steps (oil)
dt = 1.e-8 # free fall time (oil)
N = 2000 # number of steps (ball)
dt = 1.e-1 # free fall time (ball)
tf = dt*float(N) # time step [s]
vo = 0.0 # initial velocity
g = 9.8 # acceleration of gravity
boverM = 2.4e5 # linear drag coefficient
coverM = 8.7e-7 # quadratic drag coefficient
```

beta = 1.6 e-4 N/sm^2, gamma = 0.25 Ns^2/m^4
f_lin = b v = (beta D) v
f_lquad = c v^2 = (gamma D^2) v^2

baseball D = 7cm, M = 0.14 = kg, b/M = 8.0e-5,  c/M = 8.7e-3 

oildrop D = 1.5 e-6 m, M = 1e-15 kg, b/M = 2.4e5, c/M = 5.6e4

```python
def euler(vi, ti, dt, deriv): # Euler method v_i â†’ v_i+1
    vnext = vi + deriv(vi,ti) * dt
    return vnext

def freefall( vi, ti): # define differential
    #dvdt = -g - boverM*vi
    dvdt = -g - coverM*vi
    return dvdt
```

```python
t = np.linspace(0, tf, N) # time array
v = np.zeros([N]) # v0, v1, ... array

v[0] = vo # initial condition

for j in range(N-1):
    v[j+1] = euler(v[j], t[j], dt, freefall)

plt.plot(t,v,"bo")
plt.plot(t,-g*t,"r")
```
