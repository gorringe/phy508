```python
import numpy as np
import matplotlib.pyplot as plt

# example of Gauss Newton method to find value of parameter m that yields
# minimum chi^2 for chi^2 (m) = 4 m^2 (rather than chi^2 = Sum_i m x_i + c)

def f(x):
    return 4*x**2

def fprime(x):
    return 8*x
```

```python
m0 = 3 #initial guess
tolerance = 1e-3
maxtrials =  100

mi = np.zeros(maxtrials) # store descent
mi[0] = m0

for i in range(maxtrials-1):
    mi[i+1] = mi[i] - f(mi[i]) / fprime(mi[i]) # G-N eqn
    if abs( f(mi[i]) / fprime(mi[i]) ) < tolerance:
      break

chi2i = 4*mi[0:i]**2
plt.plot(mi[0:i],chi2i[0:i],"m.")
plt.plot(mi[i-1:i],chi2i[i-1:i],"go")
plt.xlabel("paramter, m")
plt.ylabel("Chi^2 (m)")
plt.title("single parameter Gauss-Newton descent")

print("iterations ", i,"best fit m ", mi[i]," tolerance ", tolerance)
```
