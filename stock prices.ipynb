```python
import numpy as np
import matplotlib.pyplot as plt

mu = 0.1         # Drift
sigma = 0.2      # Volatility
S0 = 100         # Initial price
T = 1            # Time period (1 year)
N = 252          # Number of time steps (daily)

# Generate random values for Wiener process (Brownian motion)
dt = T / N
t = np.linspace(0, T, N+1)
dW = np.random.normal(0, np.sqrt(dt), N)

# Initialize an array to store simulated stock prices
S = np.zeros(N+1)
S[0] = S0

# Simulate the stock price using the SDE
for i in range(1, N+1):
    dS = mu * S[i-1] * dt + sigma * S[i-1] * dW[i-1]
    S[i] = S[i-1] + dS

# Plot the simulated stock price
plt.figure(figsize=(10, 5))
#plt.plot(t, S)
plt.plot(t[:-1], dW)
plt.xlabel('Time (years)')
plt.ylabel('Stock Price')
plt.title('Simulated Geometric Brownian Motion')
plt.grid(True)
plt.show()
```
