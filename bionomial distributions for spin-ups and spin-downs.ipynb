```python
import numpy as np
import matplotlib.pyplot as plt

Ntrial = 100

move_n = np.arange( 0, Ntrial+1, 1) # array of possible moves
prob_n = np.zeros( Ntrial+1, dtype = float)  # array of move probabilities
for k in range(Ntrial+1):
    prob_n[k] = np.math.comb(Ntrial, k) / 2.**Ntrial  # numpy (N choose n) divided by 2^N possibilities

plt.figure(figsize=(8, 6))
plt.bar( move_n, prob_n, color="r")
plt.xlabel('# of microstates')
plt.ylabel('# of spin-ups')
```
