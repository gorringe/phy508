```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 100000 # number of random points in 1x1x1 cube
count = 0 # number of random points in '1/8th hemisphere'

a, b, c = 0.9, 0.8, 0.7
xp, yp, zp = [], [], []
xps, yps, zps = [], [], []

for j in range(N):
    x,y,z = 2.*np.random.ranf()-1.,  2.*np.random.ranf()-1.,  2.*np.random.ranf()-1.
    xp.append(x), yp.append(y), zp.append(z)
    if x**2 / a**2 + y**2 / b**2 + z**2 /c**2 < 1.:
        count += 1
        xps.append(x), yps.append(y), zps.append(z)

nvol = 2**3 * float(count) / float(N) # factor of 2**3 is volume of cube and count/N is realtive volume of sphere to cube

print( nvol, 4./3.*np.pi*a*b*c)
```

```python
ax = plt.axes(projection='3d')
ax.plot3D(xp, yp, zp, 'm.')
ax.plot3D(xps, yps, zps, 'g.')
```
