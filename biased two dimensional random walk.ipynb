```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(2101)
moves = (-1,+1) # unbiased
bias = 0.01
moves = (-1+bias,+1+bias) # unbiased
colors = ["r-","b-","g-","m-","c-","y-"]
```

```python
Nparticle = 1
Nstep = 400
step = np.arange( 0, Nstep, 1)
pathx = np.empty( Nstep)
pathy = np.empty( Nstep)
meanx = np.empty( Nstep)
meany = np.empty( Nstep)
rmsqx = np.empty( Nstep)
rmsqy = np.empty( Nstep)
pcle =  np.arange( 0, Nparticle, 1)
rngex = np.empty( Nparticle)
rngey = np.empty( Nparticle)
```

```python
for ip in range( 0 , Nparticle):
    x, y = 0., 0.
    for ixy in range( 0 , Nstep):
        pathx[ixy] = x
        pathy[ixy] = y
        #x += ( np.random.ranf() - 0.5 ) * 2. # uniform dstbn
        #y += ( np.random.ranf() - 0.5 ) * 2. # uniform dstbn
        x += np.random.choice(moves)         # move of -1 or +1
        y += np.random.choice(moves)         # move of -1 or +1
        meanx[ixy] += x
        meany[ixy] += y
        rmsqx[ixy] += x**2
        rmsqy[ixy] += y**2
    plt.plot( pathx, pathy, colors[ip%6])
    rngex[ip] = pathx[Nstep-1]
    rngey[ip] = pathy[Nstep-1]
plt.xlabel("path x-coordinate")
plt.ylabel("path y-coordinate")
meanx /= Nparticle
meany /= Nparticle
rmsqx /= Nparticle
rmsqy /= Nparticle
```

```python
plt.plot( step, meanx, "g-",label="meanx")
plt.plot( step, meany, "y-",label="meany")
plt.plot( step, rmsqx,"m-",label="rmsx")
plt.plot( step, rmsqy,"r-",label="rmsy")
plt.xlabel("step index")
plt.ylabel("mean, rms")
plt.legend()
```

```python
plt.hist( rngex, bins=40, range=[-100, 100], histtype='step',  edgecolor='red')
plt.hist( rngey, bins=40, range=[-100, 100], histtype='step',  edgecolor='blue')
plt.xlabel("x, yrange")
plt.ylabel("frequency")
```
