```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 100000
rollarray = []

for i in range(N):
    roll = np.random.randint(1, 7)
    rollarray.append(roll)
```

```python
# distribution plot of random numbers

plt.figure(figsize=(8,5))
plt.plot(rollarray,"mo")
plt.xlabel("dice roll index")
plt.ylabel("dice roll value")
```

```python
#frequency plot of random numbers
Nb = 6
yf, bins = np.histogram(rollarray,bins=Nb, range=(1.,6.))
dyf = np.sqrt(yf)
xf = np.linspace(1.,6.,Nb)
yf, dyf, dxf  = yf/N, dyf/N, 0.

plt.figure(figsize=(8,5))
plt.errorbar(xf,yf,xerr=dxf,yerr=dyf,fmt='mo')
plt.xlabel("dice roll value")
plt.ylabel("dice roll probability")
plt.ylim(0.,0.25)
```

```python
# spectral plot of roll_n versus roll_n-1

plt.figure(figsize=(8,5))
plt.plot(rollarray[1:], rollarray[:-1],"m.",markersize=8)
plt.xlabel("x_n")
plt.ylabel("x_n-1")
```

```python
# chi-squared test
ndf = Nb - 1
mean = np.mean(yf)
var = (( yf-mean) / dyf )**2
chi2 = var.sum() / ndf
print("mean, chi2, ndf ", mean, chi2, ndf)
```
