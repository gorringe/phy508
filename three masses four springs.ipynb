```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# initial position

x1 = -0.1 # freq sqrt(2)*omega_o mode
x2 = 0.0
x3 = +0.1

x1 = +0.1 # freq sqrt(2)*sqrt(1-2^(1/2))*omega_o mode
x2 = +np.sqrt(2.)*0.1
x3 = +0.1

x1 = +0.1 # freq sqrt(2)*sqrt(1+2^(1/2))*omega_o mode
x2 = -np.sqrt(2.)*0.1
x3 = +0.1

# spring constants
k1 = 1.
k2 = 1.
k3 = 1.
k4 = 1.

# initial time, velocities
t = 0
v1 = 0
v2 = 0
v3 = 0

# accelerations
a1 = -k2*(x1-x2) -k1*x1
a2 = +k2*(x1-x2) -k3*(x3-x2)
a3 = +k3*(x3-x2) -k4*x3

dt = 0.002 ## time step

#Lists for storing data
X_1 = [x1]
X_2 = [x2]
X_3 = [x3]
V_1 = [v1]
V_2 = [v2]
V_3 = [v3]
A_1 = [a1]
A_2 = [a2]
A_3 = [a3]
Time = [t]

while(t<100):
   
    # updating according to Verlet algorithm
    x1 += v1*dt/2
    x2 += v2*dt/2
    x3 += v3*dt/2
 
    a1 = -k1*x1 -k2*(x1-x2) 
    a2 = -k2*(x2-x1) -k3*(x2-x3)
    a3 = -k3*(x3-x2) -k4*x3
 
    v1 += a1*dt
    v2 += a2*dt
    v3 += a3*dt


    x1 += v1*dt/2
    x2 += v2*dt/2
    x3 += v3*dt/2

    t += dt
    
    # appen*ding lists
    X_1.append(x1)
    X_2.append(x2)
    X_3.append(x3)
    V_1.append(v1)
    V_2.append(v2)
    V_3.append(v3)
    A_1.append(a1)
    A_2.append(a2)
    A_3.append(a3)
    Time.append(t)
    
```

```python

fig,((ax1,ax2,ax3),(ax4,ax5,ax6),(ax7,ax8,ax9)) = plt.subplots(nrows = 3,ncols= 3,figsize = (18,10))

ax1.set_title(r'$X_1$ vs $t$') 
ax2.set_title(r'$X_2$ vs $t$')
ax3.set_title(r'$X_3$ vs $t$')
ax4.set_title(r'$V_1$ vs $t$')
ax5.set_title(r'$V_2$ vs $t$')
ax6.set_title(r'$V_3$ vs $t$')
ax7.set_title(r'$A_1$ vs $t$')
ax8.set_title(r'$A_2$ vs $t$')
ax9.set_title(r'$A_3$ vs $t$')
     
ax1.plot(Time,X_1,'r')
ax2.plot(Time,X_2,'g')
ax3.plot(Time,X_3,'b')
ax4.plot(Time,V_1,'r')
ax5.plot(Time,V_2,'g')
ax6.plot(Time,V_3,'b')
ax7.plot(Time,A_1,'r')
ax8.plot(Time,A_2,'g')
ax9.plot(Time,A_3,'b')

#ax4.plot(Time,np.array(X_1)-np.array(X_2),'k')
#ax5.plot(Time,np.array(X_2)-np.array(X_3),'c')
#ax6.plot(Time,np.array(X_1)-np.array(X_3),'o')
#ax4.set_ylim(-0.1,0.1)

plt.show()
```
