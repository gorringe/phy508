```python
import numpy as np
```

```python
to = input('Enter the proper time [s]: ')
to = float(to)
v = input('Enter the velocity v [c]: ')
v = float(v)
```

```python
if v >= +1.0:
  print('invalid velocity, v >= +c')
elif v <= -1.0:
  print('invalid velcity, v <= -c')
else:
  gamma = 1.0/np.sqrt(1-v*v)
  t = gamma*to
  print('Lorentz factor is :', gamma)
  print('dilated time is [s]: ', t)
```
