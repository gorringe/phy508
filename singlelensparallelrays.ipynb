```python
#FO
F1 = +0.5
F2 = -0.5
D = 0.0
L1 = 0.5
L2 = 0.5
```

```python
# define matrices
f1 = np.array([[ 1,  0] ,[ -1/F1, 1]])
f2 = np.array([[ 1,  0] ,[ -1/F2, 1]])
o1 = np.array([[ 1,  L1] ,[ 0, 1]])
o2 = np.array([[ 1,  L2] ,[ 0, 1]])
d = np.array([[ 1,  D] ,[ 0, 1]])
```

```python
#FO - focusing parallel rays
N = 10
for j in range(N): 
  r1 = np.array([[0.01*j], [0.0]])
  r2 = o2@f1@o1@r1
  print(j,r1[0], r2[0])
```
