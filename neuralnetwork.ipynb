```python


!pip install mglearn  #"Introduction to Machine Learning with Python" written by Andreas C. MÃ¼ller, Sarah Guido
import  mglearn

import matplotlib.pyplot as plt

# open-source machine learning library
from sklearn.datasets import make_moons
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
```

```python

# generates datasetcomprising two interleaving half-moon shapes, X is feature matrix, y is target variable
X, y = make_moons(n_samples=100, noise=0.2, random_state=3)

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Two-Moon-Shaped Dataset")
```

```python

# split X, y into training and testing subsets
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

# Create the first plot
plt.subplot(1, 2, 1)  # (rows, columns, panel number)
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.title("Two-Moon-Shaped Dataset - train")

# Create the second plot
plt.subplot(1, 2, 2)  # (rows, columns, panel number)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test)
plt.title("Two-Moon-Shaped Dataset - test")
```

```python
# implement multilayer perceptron classification algorithm for patterns / relationships in data
# args: hidden_layer_sizes - #nodes, #layers, activation function = relu, sigmoid, solver - optimization method, gradient descent, Quasi-Newton method
mlp = MLPClassifier(solver='lbfgs', random_state = 0).fit(X_train, y_train)

#  MÃ¼ller, Guido machine learning plots
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3) # neural network classification
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train) # training data
```
