```python
import matplotlib.pyplot as plt
import numpy as np

from google.colab import drive
drive.mount('/content/drive')
```

```python
data = np.loadtxt("/content/drive/MyDrive/Courses/phy508/sunspots.txt")

yr = data[55:320,0]
ns = data[55:320,1]

cycle = np.zeros(24)
spots = np.zeros(24)

for ic in range(0,24): 
  cycle[ic] = ic+1
  for iy in range(0,11):
     spots[ic] += ns[11*ic+iy]

print(cycle,spots)
```

```python
plt.plot(cycle,spots)
plt.title("Sunspots versus cycle")
plt.xlabel("cycle")
plt.ylabel("sunspots / cycle")
```

```python
plt.plot(yr,ns)
plt.title("Sunspots versus year")
plt.xlabel("year")
plt.ylabel("sunspot total")
```
