```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 1000
rollarray = []

for i in range(N):
    roll1 = np.random.randint(1, 7)
    roll2 = np.random.randint(1, 7)
    rollarray.append(roll1+roll2)
```

```python
# plot list of random numbers
plt.plot(rollarray,"mo")
plt.xlabel("dice roll index")
plt.ylabel("dice roll value")
```

```python
#plot frequency of randum numbers
Nb = 12
yf, bins = np.histogram(rollarray,bins=Nb, range=(1.,12.))
dyf = np.sqrt(yf)
xf = np.linspace(1.,12.,Nb)
yf, dyf, dxf = yf/N, dyf/N, 0.
plt.figure()
plt.errorbar(xf,yf,xerr=dxf,yerr=dyf,fmt='mo')
plt.xlabel("dice roll value")
plt.ylabel("dice roll probability")
plt.ylim(0.,0.25)
```
