```python
import numpy as np
```

```python
a = np.array([1,2,3,4,5,6],int)
b = np.array([[1,2,3],[4,5,6]],int)
```

```python
print(a.size, a.shape, a.dtype, a)
```

```python
print(b.size, b.shape, b.dtype, b)
```

```python
s = a[:2] # s = a[2:] # s = a[2:4] 
print(s)
#note the index starts from zero and result includes the start index, but excludes the end index.
```

```python
s = b[1,1:] # s = b[0,0:2] # s = b[0:2,0:2]
print(s)
#note the index starts from zero and result includes the start index, but excludes the end index.
```
