```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# damped oscillator
t = np.linspace(0, 5*2*np.pi, num = 100)
y = np.sin(t)*np.exp(-0.1*t)
vy = np.cos(t)*np.exp(-0.1*t)
```

```python
fig, axes = plt.subplots(2)
axes[0].plot(t,y)
axes[1].plot(t,vy)
axes[0].set_ylabel('y')
axes[1].set_ylabel('vy')
axes[1].set_xlabel('t')
fig.suptitle('position, velocity versus time')
```
