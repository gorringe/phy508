```python
import matplotlib.pyplot as plt
import scipy.optimize as fit
import numpy as np
```

```python
N = 10
m = 1.0
c = -5.0
s = 1.0
```

```python
np.random.seed(1)
xi = np.linspace( 1, N, num=N)                    # indep. variable
yi = m*xi + c                                     #   dep. variable
yi += np.random.normal( loc=0.0, scale=s, size=N) #  variable error
si = s*np.ones( (N))
plt.errorbar(xi, yi, yerr=si,fmt='mo')
plt.xlabel("x_i")
plt.ylabel("y_i")
```

```python
M = np.column_stack( (xi,np.ones(N)))             # Nx2 matrix 
MTM = np.dot( M.transpose(), M)
MTMINV = np.linalg.inv(MTM)
MTY = np.dot(M.transpose(),yi)
print(M)
print(MTM)
print(MTMINV)
print(MTY)
```

```python
pars = np.dot(MTMINV,MTY)
covs = MTMINV*s
print(pars)
print(covs)
```

```python
plt.errorbar(xi, yi, yerr=si,fmt='mo')
plt.xlabel("x_i")
plt.ylabel("y_i")
yfi =  np.dot(M,pars)
plt.plot(xi,yfi,"g-")
```

```python
meanx = np.sum( xi)
meany = np.sum( yi)
meanx2 = np.sum( xi**2)
meany2 = np.sum( yi**2)
meanxy = np.sum( xi*yi)
meanis2 = np.sum( 1./si**2)
```

```python
M = np.array([[meanx2,meanx],[meanx,meanis2]])
V = np.array([[meanxy],[meany]])
invM = np.linalg.inv(M)
pars = np.dot(invM,V)
mf = pars[0]
cf = pars[1]
dmf2 = N/(N*meanx2-meanx*meanx)
dcf2 = meanx2/(N*meanx2-meanx*meanx)
cov = -meanx/(N*meanx2-meanx*meanx)
```

```python
def func(x, a, b):
    return a*x+b
pars, covs = fit.curve_fit(func, xi, yi, sigma=si, absolute_sigma=True)
print(pars, covs)
print(mf,cf,dmf2,dcf2,cov)
```

```python
plt.errorbar(xi, yi, yerr=si,fmt='mo')
plt.xlabel("x_i")
plt.ylabel("y_i")
yfi = pars[0]*xi + pars[1]
plt.plot(xi,yfi,"g-")
#yfi = mf*xi + cf
#plt.plot(xi,yfi,"r--")
```

```python
def chisq(p0, p1, x, y, s):
    return np.sum(( (y - p0*x - p1 ) / s )**2 )
Np = 10
chi2 = np.zeros((Np, Np))
mf = np.linspace( pars[0]-3.*np.sqrt(covs[0][0]), pars[0]+3.*np.sqrt(covs[0][0]),Np)
cf = np.linspace( pars[1]-3.*np.sqrt(covs[1][1]), pars[1]+3.*np.sqrt(covs[1][1]),Np)
print(mf,cf)
im = 0
for xm in mf:
    ic = 0 
    for yc in cf:
        chi2[im][ic] = chisq(xm,yc,xi,yi,si)
        #print(xm,yc,chi2[im][ic])
        ic += 1
    im += 1 
plt.contourf(mf,cf,chi2,15)
```
