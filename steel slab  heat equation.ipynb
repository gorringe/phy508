```python
import numpy as np
import matplotlib.pylab as plt
```

```python
L = 0.01      # steel thickness [m]
D = 4.25e-6   # thermal diffusivity
N = 100       # space grid dimension
a = L/N       # space grid interval
h = 1e-4      # time grid interval

Tlo = 0.0     # low temp [deg C]
Tmi = 20.0    # int temp [deg C]
Thi = 50.0    # hgh temp [deg C]

T = np.empty(N+1,float) # temp array
T[0] = Thi
T[N] = Tlo
T[1:N] = Tmi
Tp = np.empty(N+1,float) # primed temp array
Tp[0] = Thi
Tp[N] = Tlo


t1 = 0.01
t2 = 0.10
t3 = 0.50
t4 = 1.00
t5 = 10.0
epsilon = h/1000
tend = t5 + epsilon

t = 0.0
c = h*D/(a*a)
while t<tend:

    for i in range(1,N): 
        Tp[i] = T[i] + c*(T[i+1]+T[i-1]-2*T[i])
    T,Tp = Tp,T
    t += h

    # Make plots at the given times
    if abs(t-t1)<epsilon:
        plt.plot(T,label="0.01 sec")
    if abs(t-t2)<epsilon:
        plt.plot(T,label="0.10 sec")
    if abs(t-t3)<epsilon:
        plt.plot(T,label="0.10 sec")
    if abs(t-t4)<epsilon:
        plt.plot(T,label="1.00 sec")
    if abs(t-t5)<epsilon:
        plt.plot(T,label="10.0 sec")

plt.xlabel("x")
plt.ylabel("T")
plt.legend()
```
