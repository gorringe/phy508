```python
import numpy as np
import matplotlib.pyplot as plt

Nparticle = 5000
Nstep = 200

np.random.seed(2111) # allows repeating / changing the PRNG sequence

moves = (-1,0,+1)                          # tuple of moves (immutable, faster)
colors = ["r-.","b-","g-","m-","c-","y-"] # list of colors (mutable, slower)

positn = np.arange( -Nstep, Nstep+1, 1) # location index
locatn = np.zeros( Nparticle, dtype=int) # individual particle location
distbn = np.zeros( ( Nstep, 2*Nstep+1), dtype=float) # step index versus particle spatial distribution
locatn += Nstep # start particles at location midpoint


for ix in range( 0 , Nstep): # loop over steps, index ix

  for ip in range( 0, Nparticle): # loop over particles, index ip
      distbn[ix][locatn[ip]] += 1
      dx = np.random.choice(moves) # move by -1 or +1
      locatn[ip] += dx

  if ix % 10 == 0:
    plt.plot( positn, distbn[ix,]) # plot path of each particle


plt.title("particle diffusion")
plt.xlim(-5*np.sqrt(Nstep),5*np.sqrt(Nstep))
#plt.yscale("log")
plt.xlabel("location")
plt.ylabel("number of particles")
```
