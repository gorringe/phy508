```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 1000000 # number of random points in 1x1x1 cube
count = 0 # number of random points in 'hemisphere'
xp, yp, zp = [], [], []
xps, yps, zps = [], [], []

for j in range(N):
    x,y,z = np.random.ranf(),  np.random.ranf(),  np.random.ranf()
    xp.append(x), yp.append(y), zp.append(z) 
    if x**2 + y**2 + z**2 < 1:
        count += 1
        xps.append(x), yps.append(y), zps.append(z) 

nvol = 2**3 * float(count) / float(N) 
print("numerical volume = ",nvol,"analytical volume = ",4./3.*np.pi)
```

```python
ax = plt.axes(projection='3d')
ax.plot3D(xp, yp, zp, 'm.')
ax.plot3D(xps, yps, zps, 'g.')
```
