```python
import numpy as np
```

```python
def h(T):
    G = 6.67*(10**-11) #m^3/(kg*s^2)
    M = 5.97*(10**24)  #kg
    R = 6371*(10**3)   #m
    total_height = ( (G*M*T**2) / (4.*np.pi**2) )**(1/3)
    return total_height - R

print("satelite altitude with 23.93-hour period: ", h(23.93*3600)*1.e-3, " km")
print("satelite altitude with 1-hour period: ", h(3600)*1.e-3, " km")
```

```python
def T(h):
    G = 6.67*(10**-11) #m^3/(kg*s^2)
    M = 5.97*(10**24)  #kg
    R = 6371*(10**3)   #m
    return (((4.*np.pi**2)*(R+h)**3)/(G*M))**(0.5)

print("satelite period with 0.0 m altitude : ", T(0)/3600.,  " hr")
print("satelite period with 35,800 km altitude : ", T(3.58e7)/3600.,  " hr")
```
