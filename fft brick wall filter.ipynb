```python
from scipy.fftpack import fft, rfft, irfft, fftfreq
import matplotlib.pyplot as plt
import numpy as np
```

```python
Np = 200000
tau = 70000.
omega = 1.43993e-03
cut = 0.75*omega/np.pi

x = np.zeros(Np) # generate signal in time domain
t = np.arange(0,Np,1)
for ix in range(Np):
  #x[ix] = np.exp(-ix/tau)
  x[ix] = np.sin(omega*ix)

fx = rfft(x)       # fx v f is signal in freq domain
f = fftfreq(x.size,d=t[1]-t[0])

cfx = np.copy(fx) # apply filter in freq domain
cfx[(f<cut)] = 0 

xp = irfft(fx)    
cxp = irfft(cfx)   

dc = cxp-xp # difference between filtered and unfiltered 
d = xp-x 

fx = rfft(cxp)
```

```python
fig,((ax1,ax2),(ax3,ax4),(ax5,ax6)) = plt.subplots(nrows = 3,ncols= 2,figsize = (18,10))

ax1.set_title(r'$x$ vs $t$') 
ax2.set_title(r'$fx$ vs $f$')
ax3.set_title(r'$xp$ vs $t$') 
ax4.set_title(r'$xcp$ vs $t$')
ax5.set_title(r'$xp-x$ vs $t$') 
ax6.set_title(r'$xcp-xp$ vs $t$')

ax1.plot(t,x,'r')
ax2.plot(f,fx,'g')
ax3.plot(t,xp,'r')
ax4.plot(t,cxp,'g')
ax5.plot(t,d,'r')
ax6.plot(t,dc,'g')

plt.show()
```
