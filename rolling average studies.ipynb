```python
import matplotlib.pyplot as plt
import scipy.optimize as fit
import scipy.signal as sg
import scipy.fft as ft
import numpy as np
```

```python
N = 1000            # time range (sec)
dt = 1.0               # time interval (sec)
amp = 1.0              # amplitude
omega = 2.*np.pi*0.05  # frequency
phi = 0.0              # phase
```

```python
taxis = np.linspace( 1, N, num = N, dtype=int )
xi = np.zeros( (N), dtype=float)
for i in range(N):
    xi[i] = amp*np.sin(omega*i+phi)

plt.plot(xi)
plt.xlabel("time (sec)")
plt.ylabel("ampl.")
plt.savefig('x0.05.png')
```

```python
fxi = ft.fft(xi)[0:N//2] 
faxis = ft.fftfreq(N,1)[0:N//2]
plt.plot(faxis,np.abs(fxi))
plt.xlabel("freq. (Hz)")
plt.ylabel("ampl.")
plt.savefig('f0.05.png')
```

```python
Nw = 200
gain = np.zeros( (Nw), dtype=float)
wdth = np.zeros( (Nw), dtype=float)

for iw in range(1,Nw):
    win = np.repeat([1],iw)
    yi = sg.convolve( xi, win, mode="same") / Nw
    gain[iw] = np.amax(yi[250:750])
    wdth[iw] = iw
    
plt.xlabel("time")
plt.ylabel("ampl.")
plt.plot(wdth,gain)
plt.savefig('g0.05.png')
```
