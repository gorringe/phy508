```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def z(n, c):
    z = 0
    for i in range(n-1):
        z = z**2 + c
    return z

#for n in range(10):
#    print(f"z({n}) = {z(n, c=+1)}") # f-string encloses variables in {} # try c = +1, 0, -1
```

```python
def complex_matrix(xmin, xmax, ymin, ymax, pixel_density):
    re = np.linspace(xmin, xmax, int((xmax - xmin) * pixel_density))
    im = np.linspace(ymin, ymax, int((ymax - ymin) * pixel_density))
    return re[np.newaxis, :] + im[:, np.newaxis] * 1j

n_interations = 21
complex_mat = complex_matrix(-2,2,-2,2,200)
binary_mat = np.where( np.abs( z( n_interations, complex_mat))  < 2, 1, 0)

plt.imshow( binary_mat)
```
