```python
import numpy as np
```

```python
x = [] # declare list with zero elements
n = int( input("# of measurements "))
```

```python
i = 0
while i < n:
    xin = float( input("enter measurement "))
    x.append(xin)
    i += 1
```

```python
avg = sum(x)/len(x)
```

```python
def variance(x):
    var = 0
    for xi in x:
        var += (xi-avg)**2
    var /= len(x) - 1
    return var
```

```python
var = variance(x)
sgma = np.sqrt(var)
```

```python
print("data", x, "min", min(x), "max", max(x), "avg", avg, "sgma", sgma)
```
