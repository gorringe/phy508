```python
import numpy as np
import matplotlib.pylab as plt
```

```python
L = 0.10      # steel thickness [m]
D = 4.25e-6   # thermal diffusivity
N = 100       # grid size
a = L/N       # grid spacing
h = 1e-2      # time step

Tlo = 0.0     # low temp [deg C]
Tmi = 100.0    # int temp [deg C]
Thi = 0.0    # hgh temp [deg C]

t1 = 1.
t2 = 10.
t3 = 100.
t4 = 500.
t5 = 1000.
epsilon = h/1000
tend = t5 + epsilon

T = np.empty(N+1,float) # temp array
T[0] = Thi
T[N] = Tlo
T[1:N] = Tmi
Tp = np.empty(N+1,float) # primed temp array
Tp[0] = Thi
Tp[N] = Tlo

t = 0.0
c = h*D/(a*a)
while t<tend:

    for i in range(1,N): 
        Tp[i] = T[i] + c*(T[i+1]+T[i-1]-2*T[i])
    T,Tp = Tp,T
    t += h

    # Make plots at the given times
    if abs(t-t1)<epsilon:
        plt.plot(T,label="1.0 sec")
    if abs(t-t2)<epsilon:
        plt.plot(T,label="10. sec")
    if abs(t-t3)<epsilon:
        plt.plot(T,label="100. sec")
    if abs(t-t4)<epsilon:
        plt.plot(T,label="500. sec")
    if abs(t-t5)<epsilon:
        plt.plot(T,label="1000. sec")

plt.xlabel("position, x")
plt.ylabel("temperature, T")
plt.legend()
```
