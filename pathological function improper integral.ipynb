```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin( 1./(x*(2-x)) )**2

def g(x):
    return 1/(x*(2-x))
```

```python
x = np.linspace(0.,2.,1000)
y1 = f(x)
y2 = g(x)

plt.semilogy(x,y1,"m")
plt.semilogy(x,y2,"r")
```

```python
N = 10000
count = 0

for i in range(N):
    xr = 2*np.random.ranf()
    yr = np.random.ranf()
    if yr < f(xr):
       count += 1
I = 2*count/N # rectangle has area 2
print(I)
```
