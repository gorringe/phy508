```python
from scipy.fft import fft, fftfreq
import matplotlib.pyplot as plt
import numpy as np
from google.colab import drive

drive.mount('/content/drive')
```

```python
data = np.loadtxt("/content/drive/MyDrive/Courses/phy508/sunspots.txt")

yr = data[:,0]
ns = data[:,1]
#yr = data[140:,0]
#ns = data[140:,1]
Np = len(ns)

sf = fft(ns)[1:Np//2]      # get fft & remove mirror 0 ... Np/2-1, Np/2 ... Np-1
f = fftfreq(Np,1)[1:Np//2] # args. are sample number, sample dt, & remove mirror
T=1/f

fig, ax = plt.subplots(3,figsize=(12,8))

ax[0].plot(yr,ns,'r')
ax[0].set_title("Sunspots numbers, frequencies")
ax[0].set_ylabel("sunspot number per year, ns")

ax[1].plot(f, np.abs(sf),'b')   # magnitude of complex fy
ax[1].set_ylabel("sunspot cycles per year, f")
#ax[1].set_xlim(0.0,0.3)

ax[2].plot(T, np.abs(sf),'g')   # magnitude of complex fy
ax[2].set_ylabel("sunspot period in years, T")
#ax[2].set_xlim(0.0,200.)
```
