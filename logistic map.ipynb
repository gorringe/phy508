```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic(r, x):
    return r * x * (1 - x)
```

```python
generations = 100
xi = np.zeros(generations)
r = 3.7
xi[0] = 0.02

for i in range(generations-1):
    xi[i+1]  = logistic(r,xi[i])

plt.plot(xi)
plt.title('population versus generation, r = 3.5699, xo =0.02')
plt.xlabel('generation')
plt.ylabel('population')
```

```python
r = 3.82
xi[0] = 0.02

for i in range(generations-1):
    xi[i+1]  = logistic(r,xi[i])
plt.plot(xi,'g')

r = 3.820000001
xi[0] = 0.02

for i in range(generations-1):
    xi[i+1]  = logistic(r,xi[i])
plt.plot(xi,'r')
```
