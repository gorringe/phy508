```python
import numpy as np
import matplotlib.pyplot as plt

N = 100
dt = 1
T = 25.0

# Generate the sine wave
t = np.linspace(0, N, int(N/dt))
y = np.cos(2 * np.pi * t / T)

# Plot the histogram
plt.figure(figsize=(10, 5))
plt.plot(t,y)
plt.xlabel('time, t')
plt.ylabel('y(t)')
plt.show()

# Calculate the DFT of the histogram
fy = np.fft.fft(y)[0:int(N/2)]
f = np.fft.fftfreq(N, dt)[0:int(N/2)]
magn = np.abs(fy)
phas = np.angle(fy)

# Plot the histogram of the DFT
plt.figure(figsize=(10, 5))
plt.plot(f,magn,".")
plt.xlabel('frequency (t^-1)')
plt.ylabel('magnitude f(y)')
plt.show()

# Plot the histogram of the DFT
plt.figure(figsize=(10, 5))
plt.plot(f,phas,".")
plt.xlabel('frequency (t^-1)')
plt.ylabel('phase f(y)')
plt.show()
```
