```python
import scipy.interpolate as interpolate
import matplotlib.pyplot as plt
import numpy.linalg as linalg
import numpy as np
```

```python
# define data points (xi, yi), declare interpolating function (x,y)

N = 2
xi = np.array([2,5])      # independent data points
yi = np.array([1.5,4.0])  # dependent data points

x = np.linspace(xi[0], xi[-1], num=50, dtype=float) # interpolating function indep variable
y = np.zeros((50), dtype=float)  # interpolating function dep variable

plt. plot(xi,yi,'mo')
```

```python
# solve for interpolating function using matrix method

M = np.column_stack( (xi,np.ones(N))) # create Vandermonde matrix

a = linalg.solve(M, yi) # solve M.a = y for a knowing vector y & matrix m
print("M.a = ", np.dot(M,a), "y = ", yi) # demo that M.a = y for vector a from solve()

y = a[0]*x+a[1]  # contruct interpolating polynomial
plt.plot( xi, yi, "mo", label='data')
plt.plot( x, y, "g-", label='function')
plt.legend()
```

```python
# solve for interpolatiing function using lagrange polynomials

lp0 = ( x-xi[1] ) / ( (xi[0]-xi[1]) ) # LP0 = (x-x1) / ( (x0-x1) ) lagrange basis polynomial c_0, i.e. i=0, m=1
lp1 = ( x-xi[0] ) / ( (xi[1]-xi[0]) ) # LP1 = (x-x0) / ( (x1-x0) ) lagrange basis polynomial c_1, i.e. i=1, m=0
# lp0 is 1 at xi[0], and 0 at xi[1] ( similarily lp1)

y = lp1*yi[1] + lp0*yi[0] # contruct interpolating polynomial
plt.plot( xi, yi, "mo", label='data')
plt.plot( x, y, "g-", label='function')
plt.plot( x, lp0, "r-", label='lp0')
plt.plot( x, lp1, "b-", label='lp1')
plt.legend()
```

```python
# solve for interpolatiing function using scipy lagrange interpolation package

poly = interpolate.lagrange(xi, yi)  # returns a numpy polynomial object
print(poly)               # print polynomial object
print(poly(2.), poly(5.)) # evaluate polynomial object

plt.plot( xi, yi, "mo", label='data')
plt.plot( x, poly(x), "g-", label='function')
plt.legend()
```
