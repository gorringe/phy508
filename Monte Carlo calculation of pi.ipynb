```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 1000000 # number of random points in 1x1x1 cube
count = 0 # number of ransom points in hemisphere
xp, yp = [], []
xpc, ypc = [], []

for j in range(N):
    x,y = np.random.ranf(),  np.random.ranf()
    xp.append(x), yp.append(y)
    if x**2 + y**2 < 1:
        count += 1
        xpc.append(x), ypc.append(y) 

Pi = 4.*float(count)/float(N)
print("numpy pi = %16.8f, ,user pi = %16.8f" % ( np.pi, Pi))
```

```python
plt.scatter(xp,yp,color="magenta")
plt.scatter(xpc,ypc,color="green")       
```
