```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic(r, x):
    return r * x * (1 - x)
```

```python
# define array of starting populations
r = np.linspace(0, 4, num=100) #
x = 0.5*np.ones(100)          # 0.5 starting value for initial population

plt.plot(r,x,"b.")
plt.title("initial population versus reproduction rate")
plt.xlabel("reproduction rate, r")
plt.ylabel("initial/maximum population, x")
```

```python
#let population evolve one generation
x  = logistic(r,x)

plt.plot(r,x,"b.")
plt.title("one-gen population versus reproduction rate")
plt.xlabel("reproduction rate, r")
plt.ylabel("one-gen/maximum population, x")
```

```python
# let population evolve many generations
iterations = 1000
for i in range(iterations):
    x  = logistic(r,x)

plt.plot(r,x,"b.")
plt.title("one-gen population versus reproduction rate")
plt.xlabel("reproduction rate, r")
plt.ylabel("one-gen/maximum population, x")
```

```python
iterations = 100
for i in range(iterations):
    x  = logistic(r,x)
    plt.scatter(r,x,c="b",marker=".")


plt.xlabel("reproduction rate, r")
plt.ylabel("actual/maximum population, x")
```
