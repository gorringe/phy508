```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic(r, x):
    return x + r * x * (1 - x)
    #return (r+1) * x * (1 - x)
```

```python
generations = 100
xi = np.zeros(generations)
r = 0.7
xi[0] = 0.05

for i in range(generations-1):
    xi[i+1]  = logistic(r,xi[i])
plt.plot(xi)
```
