
fast algorithm for primes based on:

i) number n is prime if having no prime factors < n. 
i.e. only need to check if n is divisible by primes

ii) non-primes n have prime factors <= √n
i.e. only need to check for prime factors up to √n.

iii) if we find a prime factor  <= √n the number 
is non-prime

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
primes = [2]
N = 100
Nprime = 1

for n in range(primes[0]+1,N+1):
    i=0
    isPrime = True
    while primes[i] <= np.sqrt(n):
        if (n%primes[i]) == 0:
            isPrime = False
            break
        i += 1
    if isPrime == True: 
      primes.append(n)
      Nprime += 1

print(Nprime, primes)
```
