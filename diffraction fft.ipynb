```python
from scipy.fft import fft, fftfreq
import matplotlib.pyplot as plt
import numpy as np
```

```python
Np = 1000
width = 100
x = np.zeros(Np)
for ix in range(Np):
  if ix < width:
    x[ix] = 1

fx = fft(x)[1:Np//2]       # get fft & remove mirror 0 ... Np/2-1, Np/2 ... Np-1
xfx = fftfreq(Np,1)[1:Np//2] # arguments are sample number, sample dt, &  remove mirror
plt.title("Sunspot cycle frequencies")
plt.plot(xfx, np.abs(fx))   # magnitude of complex fy
plt.xlabel("sunspot cycles per year")
plt.xlim(0.0,0.3)
```
