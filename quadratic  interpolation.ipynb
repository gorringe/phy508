```python
import scipy.interpolate as interpolate
import matplotlib.pyplot as plt
import numpy.linalg as linalg
import numpy as np
```

```python
# define data points (xi, yi), declare interpolating function (x,y)

N = 3
xi = np.array([1.,2.,3.])      # independent data points
yi = np.array([1,4.,9.])  # dependent data data points

x = np.linspace(xi[0], xi[-1], num=50, dtype=float) # interpolating function indep variable
y = np.zeros((50), dtype=float)  # interpolating function dep variable

plt. plot(xi,yi,'mo')
```

```python
# solve for interpolating function using matrix method

M = np.column_stack( ( xi*xi, xi, np.ones(N))) # Vandermonde matrix
                             # (geometric progression)

a = linalg.solve(M, yi) # solves M a = y for vectors a,y, matrix m
print("M.a = ", np.dot(M,a), "y = ", yi)


y = a[0]*x*x+a[1]*x+a[2]  # contruct interpolating polynomial
plt.plot( xi, yi, "mo", label='data')
plt.plot( x, y, "g-", label='function')
plt.legend()
```

```python
# solve for interpolatiing function using lagrange polynomials

lp0 = ( x-xi[1] ) / ( (xi[0]-xi[1]) ) * ( x-xi[2] ) / ( (xi[0]-xi[2]) ) # LP0 = (x-x1) / ( (x0-x1) ) lagrange polynomial c_0, i.e. i=0, m=1,2
lp1 = ( x-xi[0] ) / ( (xi[1]-xi[0]) ) * ( x-xi[2] ) / ( (xi[1]-xi[2]) ) # LP1 = (x-x0) / ( (x1-x0) ) lagrange polynomial c_1, i.e. i=1, m=0,2
lp2 = ( x-xi[0] ) / ( (xi[2]-xi[0]) ) * ( x-xi[1] ) / ( (xi[2]-xi[1]) ) # LP1 = (x-x0) / ( (x1-x0) ) lagrange polynomial c_1, i.e. i=2, m=0,1

y = lp2*yi[2] + lp1*yi[1] + lp0*yi[0] # contruct interpolating polynomial
plt.plot( xi, yi, "mo", label='data')
plt.plot( x, y, "g-", label='function')
plt.legend()
```

```python
# solve for interpolatiing function using scipy lagrange interpolation package

poly = interpolate.lagrange(xi, yi)  # returns a numpy polynomial object
print(poly)                          # print polynomial
print(poly(2.), poly(5.))            # evaluate polynomial

y = poly(x)
plt.plot( xi, yi, "mo", label='data')
plt.plot( x, y, "g-", label='function')
plt.legend()
```
