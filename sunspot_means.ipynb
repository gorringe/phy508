```python
import matplotlib.pyplot as plt
import numpy as np

from google.colab import drive
drive.mount('/content/drive')
```

```python
data = np.loadtxt("/content/drive/MyDrive/Courses/phy508/sunspots.txt")

yr = data[55:320,0]
ns = data[55:320,1]

cycle = np.zeros(24)
avg = np.zeros(24)
var =  np.zeros(24)

print(" cycle mean variance")

for ic in range(0,24): 

  cycle[ic] = ic+1

  for iy in range(0,11):
    avg[ic] += ns[11*ic+iy]
  avg[ic] /= 11.

  for iy in range(0,11):
    var[ic] += ( ns[11*ic+iy] - avg[ic] )**2
  var[ic] /= 10.

  print(" %2d %7.2f %8.2f" % (cycle[ic], avg[ic], var[ic]))
```

```python
plt.plot(cycle,avg, "m")
plt.title("Sunspot mean versus cycle")
plt.xlabel("cycle")
plt.ylabel("sunspot mean / cycle")
```

```python
plt.plot(yr,ns)
plt.title("Sunspots versus year")
plt.xlabel("year")
plt.ylabel("sunspot total")
```
