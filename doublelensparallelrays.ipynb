```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#FODO
F1 = +0.5
F2 = -0.5
D = 0.1
L1 = 2.5+F1
L2 = 2.5
```

```python
# define matrices
f1 = np.array([[ 1,  0] ,[ -1/F1, 1]])
f2 = np.array([[ 1,  0] ,[ -1/F2, 1]])
o1 = np.array([[ 1,  L1] ,[ 0, 1]])
o2 = np.array([[ 1,  L2] ,[ 0, 1]])
d = np.array([[ 1,  D] ,[ 0, 1]])
```

```python
#FODO
N = 10
for j in range(N): 
  #r1 = np.array([[0.001*j], [0.0]])
  r1 = np.array([[0.0], [0.001*j]])
  #r2 = o2@f2@d@f1@o1@r1
  r2 = o1@r1
  r3 = f1@r2
  r4 = d@r3
  r5 = f2@r4
  r6 = o2@r5
  plt.scatter(r1[1],r1[0],s=10.0)
  #plt.scatter(r2[1],r2[0])
  plt.scatter(r3[1],r3[0],s=20.0)
  #plt.scatter(r4[1],r4[0])
  plt.scatter(r5[1],r5[0],s=30.0)
  plt.scatter(r6[1],r6[0],s=40.0)

```
